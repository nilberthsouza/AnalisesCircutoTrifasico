```python
import numpy as np
import matplotlib.pyplot as plt
```

# Tempo para simulação


```python
t = np.linspace(0, 1/f, 1000)
```

# Definições iniciais


```python

V_magnitude = 127  # Tensão de linha (Vrms)
f = 60
omega = 2 * np.pi * f  # 377
R = 45
C = 41.1e-6
Zc = -1j / (omega * C)

```

```python
# Impedância total
Z_total = R + Zc

```

# Sequência de fase positiva


```python

Va_phase = 0
Vb_phase = -120
Vc_phase = 120  #

```

```python


# Tensões de fase
Va = V_magnitude * np.sqrt(2) * np.cos(omega * t + np.deg2rad(Va_phase))
Vb = V_magnitude * np.sqrt(2) * np.cos(omega * t + np.deg2rad(Vb_phase))
Vc = V_magnitude * np.sqrt(2) * np.cos(omega * t + np.deg2rad(Vc_phase))

# Correntes de fase
Ia = Va / Z_total
Ib = Vb / Z_total
Ic = Vc / Z_total

# Tensões de linha
Vab = Va - Vb
Vbc = Vb - Vc
Vca = Vc - Va

```

# Potências Aparentes


```python
S_a = Va * np.conj(Ia)
S_b = Vb * np.conj(Ib)
S_c = Vc * np.conj(Ic)
S_total = S_a + S_b + S_c  # Potência total trifásica

P_ativa = S_total.real
Q_reativa = S_total.imag
fp = P_ativa / abs(S_total)
```

# Gráficos das tensões de fase


```python

plt.figure(figsize=(12, 6))
plt.plot(t, Va, 'r', label="Va (Vermelho)")
plt.plot(t, Vb, 'g', label="Vb (Verde)")
plt.plot(t, Vc, 'b', label="Vc (Azul)")
plt.title("Tensões de Fase")
plt.xlabel("Tempo (s)")
plt.ylabel("Tensão (V)")
plt.legend()
plt.grid()
plt.show()
```

# Gráficos das correntes de fase


```python

plt.figure(figsize=(12, 6))
plt.plot(t, Ia.real, 'r', label="Ia (Vermelho)")
plt.plot(t, Ib.real, 'g', label="Ib (Verde)")
plt.plot(t, Ic.real, 'b', label="Ic (Azul)")
plt.title("Correntes de Fase")
plt.xlabel("Tempo (s)")
plt.ylabel("Corrente (A)")
plt.legend()
plt.grid()
plt.show()
```

# Gráficos das tensões de linha


```python

plt.figure(figsize=(12, 6))
plt.plot(t, Vab, 'r', label="Vab (Vermelho)")
plt.plot(t, Vbc, 'g', label="Vbc (Verde)")
plt.plot(t, Vca, 'b', label="Vca (Azul)")
plt.title("Tensões de Linha")
plt.xlabel("Tempo (s)")
plt.ylabel("Tensão (V)")
plt.legend()
plt.grid()
plt.show()
```
